# Understanding Ratios Odds and Currencies

## 1] Odds

A classical range is defined by counting observations and constraining the proportional probability to be between 0-1.

Odds is defined as a ratio such as 2:1, where we say we are twice as likely against 1:1.

Likelihood is a percentage where 100% is equal chances. It is just the the odds ratio as a percentage.

Log odds constrains the odds ratio between -2 and 2 in a log curved consistant way.

![title](img/Odds_spectral_table.png)

## 2] Log Odds Compare and Contrast

```python
# Sample Data
# Metric Odds VS Log Odds VS Proportions
SAMPLE = {
    'Numerator': [
        1, 1, 1, 1, 49, 1, 51, 2, 4, 5, 6
    ],
    'Denominator': [
        6, 5, 4, 2, 50, 1, 50, 1, 1, 1, 1
    ]
}
import pandas as pd
import math
df = pd.DataFrame(SAMPLE)
df['MetricOdds'] = df['Numerator'] / df['Denominator']
df['ClassicalRange'] = df['Numerator'] / (
    df['Denominator'] + df['Numerator']
)
df['LogOdds'] = df['MetricOdds'].apply(
    lambda x: math.log(x)
)
df[[
    'MetricOdds', 'LogOdds', 'ClassicalRange'
]].plot(
    marker='o', grid=True,
    xticks=[0, 5, 10],
    title=(
        'Metric Odds VS Log Odds VS ' +
        'Proportional Classical Range'
    ),
    ylabel='Metric Ratio Value',
    xlabel='Ratio Example Weakest to Strongest'
)
df[['MetricOdds', 'LogOdds', 'ClassicalRange']]
```

## 3] Currencies

The numerator and denomintator of a ratio does not have to be an observational count. It can represent other types of value like currency.

The currency rate, is the odds in percent as a decimal. This means a rate of 1 is equal strength at 100%.

![title](img/Currencies_spectral_table.png)

Investopedia - https://www.investopedia.com/terms/q/quotecurrency.asp - 
"Let's assume a trader wants to purchase 400GBP using U.S. dollars. This would involve a trade using the GBP/USD currency pair. To execute the trade, they need to figure out how many USD (the quote currency) they need to sell to get 400GBP. 

If the exchange rate for the pair at the end of the trading day is 1.4103, it costs the trader 1.4103USD to purchase 1GBP. To complete the transaction on that day, the trader has to sell 564.12 units of the quote currency to get 400 units of the base currency or 564.12USD for 400GBP = (400 x 1.4103)."

```python
# 1GBP is 1.4103USD
GBPUSD_rate = 1.4103

# I want to buy 400GBP
USD_to_sell = 400 * GBPUSD_rate
USD_to_sell
```

In currencies, it is important to remember that the denominitor is the base. It is always 1.

## 4] Ratio Behaviours

```python
# Sample Data - Rising Numerator
SAMPLE = {
    'Numerator': [
        1, 2, 3, 4, 3, 6, 7, 8, 9, 10
    ],
    'Denominator':  [
        10, 10, 10, 10, 10, 10, 10, 10, 10, 10
    ]
}
import pandas as pd
df = pd.DataFrame(SAMPLE)
df['RatioScenario'] = \
    df['Numerator'] / df['Denominator']
df['RatioScenario'].plot(grid=True, marker='o')
```
